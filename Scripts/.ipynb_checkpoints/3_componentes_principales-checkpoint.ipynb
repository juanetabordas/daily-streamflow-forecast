```python
import pandas as pd
import numpy as np
import os
import glob
import seaborn as sns
import xarray as xr
import matplotlib.pyplot as plt
import datetime as dt

from PIL import Image
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

from Funciones import *
```

```python
ruta_base  = os.path.dirname(os.path.abspath('./')) + '/'
ruta_datos = ruta_base + 'Datos/'
ruta_shp   = ruta_datos+ 'Shapefiles/'
```

```python
# Carga información y elimina datos nans

datos_anomalias_rez = pd.read_csv(ruta_datos+'Anomalias_rezagadas_datos.csv', index_col=0, parse_dates=True)
datos_anomalias_rez = datos_anomalias_rez.iloc[~datos_anomalias_rez.isna().any(axis=1).values,:]
```

### Definición de variables independientes y dependiente

```python
X = datos_anomalias_rez.drop(['Caudal'],axis=1)
y = datos_anomalias_rez['Caudal']
```

### Aplicar PCA con scikit-learn

```python
n = X.shape[1]  # número de features
pca = PCA(n_components=n).fit(X)
print(pca.components_.shape)
```

```python
print(pca.explained_variance_ratio_)
```

```python
plt.scatter(np.arange(1, X.shape[1]+1,1), pca.explained_variance_ratio_)
plt.xlabel('Number of components')
plt.ylabel('Explained variance')
plt.show()
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
pca_samples = pca.transform(X)
```

```python
pca_samples
```

```python
### Relación entre las primeras PCs y las variables
```

```python
xvector = pca.components_[0] 
yvector = pca.components_[1]

xs = pca.transform(X)[:,0] 
ys = pca.transform(X)[:,1]

## Para visualizar las proyecciones de cada variable en los componentes se utiliza la siguiente función
for i in range(len(xvector)):
# arrows project features (ie columns from csv) as vectors onto PC axes
    plt.arrow(0, 0, xvector[i]*max(xs), yvector[i]*max(ys),
              color='r', width=0.0005, head_width=0.0025)
    plt.text(xvector[i]*max(xs)*1.2, yvector[i]*max(ys)*1.2,
             list(X.columns.values)[i], color='r') 

# for c, i, iris.target_names in zip("rgb", [0, 1, 2], iris.feature_names):
#     plt.scatter(pca_samples[y==i,0], pca_samples[y==i,1], c=c)
plt.tick_params('y', colors='k', labelsize=12, length=2)
plt.tick_params('x', colors='k', labelsize= 12, length=2)
plt.xlabel("Componente Principal 1", fontsize=16)
plt.ylabel("Componente Principal 2", fontsize=16)
# plt.legend(iris.feature_names, fontsize=8)
plt.show()
```
